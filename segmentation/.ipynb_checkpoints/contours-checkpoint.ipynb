Contours:

```python
import cv2
import numpy as np
import urllib

# image = cv2.imread('/home/nikamanth/pythondir/cv/index.jpeg')

req = urllib.urlopen('https://vignette.wikia.nocookie.net/verse-and-dimensions/images/5/53/Reccc-0.png/revision/latest?cb=20180201225606')
arr = np.asarray(bytearray(req.read()), dtype=np.uint8)
image = cv2.imdecode(arr, -1)
image = cv2.pyrDown(image)

#converting to gray scale
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
cv2.imshow('gray',gray)
cv2.waitKey(0)

#bluring for noise reduction -- nice method
blur = cv2.GaussianBlur(gray,(5,5),0)
cv2.imshow('blur',blur)
cv2.waitKey(0)

#canny for edge detection
edges = cv2.Canny(blur,30,70)
cv2.imshow('canny',edges)
cv2.waitKey(0)

#finding contour
contouredimg = edges.copy()
im2, contours, heirarchy = cv2.findContours(contouredimg, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
cv2.imshow('contours',contouredimg)
cv2.waitKey(0)

print "no of contours found:"+str(len(contours))
#drawing contour

all = image.copy()
cv2.drawContours(all,contours,-1,(211,0,148),3)
cv2.imshow('contoured image',all)
cv2.waitKey(0)

cv2.destroyAllWindows()
```

Sorting: by area 
//run above code first

```python
def get_area(contours):
    a = []
    for contour in contours:
        area = cv2.contourArea(contour)
        a.append(area)
    return a

# before sorting
print get_area(contours)

#sorting
sorted_contours = sorted(contours, key=cv2.contourArea, reverse=True)

# after sorting
print get_area(sorted_contours)

for contourobject in sorted_contours:
    one = image.copy()
    cv2.drawContours(one,[contourobject],0,(211,0,148),3)
    cv2.imshow('contoured image',one)
    cv2.waitKey(0)
cv2.destroyAllWindows()
```

Approximating contours

```python
import numpy as np
import cv2
import urllib

#Loading image
image = cv2.imread('/home/nikamanth/Documents/mynotes/cv/segmentation/house.png')
org = image.copy()

# req = urllib.urlopen('https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcRsoZAj_1RGrn5rXS-DQMQow3OCrCpEbSxX5Krj8Gwr6BekSUe2wA')
# arr = np.asarray(bytearray(req.read()), dtype=np.uint8)
# image = cv2.imdecode(arr, -1)
# image = cv2.pyrUp(image)


#gray scaling the image
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
_,thresh = cv2.threshold(gray,127,255,cv2.THRESH_BINARY + cv2.THRESH_OTSU)

#finding contour
contouredimg = thresh.copy()
im2, contours, heirarchy = cv2.findContours(contouredimg, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
cv2.imshow('contours',contouredimg)
cv2.waitKey(0)

n = len(contours) - 1
contours = sorted(contours, key = cv2.contourArea, reverse = False)[:n]
print len(contours)

for contour in contours:
    x,y,w,h = cv2.boundingRect(contour)
    cv2.rectangle(image,(x,y),(x+w,y+h),(255,25,0),2)
    cv2.imshow("bounding rectangle",image)
    cv2.waitKey(0)
    
for contour in contours:
    accuracy = 0.035*cv2.arcLength(contour, True)
    print accuracy
    approx = cv2.approxPolyDP(contour,accuracy,True)
    print len(approx) #gives the no of sides
    cv2.drawContours(org,[approx],0,(211,0,148),3)  
    cv2.imshow('approx',org)
    cv2.waitKey(0)
    
#cv2.imshow("house",thresh)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

convex hull:

```python
import numpy as np
import cv2
import urllib

#Loading image
image = cv2.imread('/home/nikamanth/Documents/mynotes/cv/segmentation/hand.jpg')
image = cv2.pyrDown(image)

org = image.copy()
cv2.imshow('original image',org)
cv2.waitKey(0)

#gray scaling the image
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
_,thresh = cv2.threshold(gray,200,255,0)
cv2.imshow('thresh',thresh)
cv2.waitKey(0)

#finding contour
contouredimg = thresh.copy()
im2, contours, heirarchy = cv2.findContours(contouredimg, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)

n = len(contours) - 1
contours = sorted(contours, key = cv2.contourArea, reverse = False)[:n]
print len(contours)

for contour in contours:
    hull = cv2.convexHull(contour)
    cv2.drawContours(org,[hull],0,(211,0,148),3)  
cv2.imshow('approx',org)
cv2.waitKey(0)

cv2.destroyAllWindows()
```
